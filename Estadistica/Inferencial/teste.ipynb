```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap

# Datos de ejemplo
x = np.random.rand(50) * 100
y = np.random.rand(50) * 100
z = np.random.rand(50) * 10  # Valores para colorear

# Crear un colormap discreto con 20 colores
cmap = ListedColormap(plt.cm.viridis(np.linspace(0, 1, 30)))

# Crear figura y ajustar tamao
plt.figure(figsize=(12, 10))

# Crear grfico de dispersin con colormap discreto
scatter = plt.scatter(x, y, c=z, cmap=cmap, vmin=0, vmax=10)

# Aadir una barra de colores para ver el mapeo de colores
plt.colorbar(scatter, label='Valores de z')

# Aadir etiquetas y ttulo
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Grfico de Dispersin con Colormap Discreto')

# Aadir cuadrcula
plt.grid(True)

# Mostrar el grfico
plt.show()
```
