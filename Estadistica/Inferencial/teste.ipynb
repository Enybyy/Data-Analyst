```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap

# Datos de ejemplo
x = np.random.rand(50) * 100
y = np.random.rand(50) * 100
z = np.random.rand(50) * 10  # Valores para colorear

# Crear un colormap discreto con 20 colores
cmap = ListedColormap(plt.cm.viridis(np.linspace(0, 1, 30)))

# Crear figura y ajustar tamaño
plt.figure(figsize=(12, 10))

# Crear gráfico de dispersión con colormap discreto
scatter = plt.scatter(x, y, c=z, cmap=cmap, vmin=0, vmax=10)

# Añadir una barra de colores para ver el mapeo de colores
plt.colorbar(scatter, label='Valores de z')

# Añadir etiquetas y título
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de Dispersión con Colormap Discreto')

# Añadir cuadrícula
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
