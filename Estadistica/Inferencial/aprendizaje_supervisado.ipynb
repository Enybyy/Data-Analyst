#### Librerías Importadas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

#### Leer data

```python
df = pd.read_csv("db/datos_correlacion.txt")
df
```

```python
# Variables de cada serie del df
x = df['X']
y = df['Y']
```

#### Calcular el coeficiente de correlación de Pearson entre las variables

```python
# Coeficiente de correlación de Pearson
np.corrcoef(x,y)
```

#### Calcular el coeficiente de correlación de Pearson entre las variables y valor-p

```python
stats.pearsonr(x,y)
```

```python
plt.figure(figsize=(10,8))

plt.scatter(x, y, 
            c=y,  
            s=50, 
            edgecolors='blue', 
            linewidths=0.5, 
            cmap='viridis', 
            vmin=0, 
            vmax=9)

plt.title("Diagrama de correlación")
```
