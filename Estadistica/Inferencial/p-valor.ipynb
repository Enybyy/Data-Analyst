#### Librerias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

np.random.seed(42)
```

#### Nuevo Medicamento

```python
df_medicametos = pd.read_csv("db/medicamentos.csv")
df_medicametos.head()
```

```python
df_medicametos_muestra = df_medicametos[:150].copy()
```

#### Nivel de significación 0.05

### Realizar el siguiente contraste de hipótesis


## CONTRASTE UNILATERAL DERECHO

```python
media_muestral = df_medicametos_muestra['RES'].mean()
```

```python
df_medicametos_muestra2 = df_medicametos_muestra['RES'] - media_muestral
round(df_medicametos_muestra2.mean(),5)
```

```python
boot_medias = []

for _ in range(10000):
    df_boot = df_medicametos_muestra2.sample(150, replace=True)
    boot_medias.append(df_boot.mean())
```

```python
plt.hist(boot_medias);
```

```python
media_muestral
```

### Probabilidad de obtener los datos

```python
print(f"La probabilidad de obtener los datos dado la hipotesis nula es de: {(boot_medias > media_muestral).mean()}")
```

## CONTRASTE UNILATERAL IZQUIERDO

```python
print(f"La probabilidad de obtener los datos dado la hipotesis nula es de: {(boot_medias <= media_muestral).mean()}")
```

## CONTRASTE BILATERAL

#### Contraste e hipótesis con un nivel de significación de a=0.05

```python
print(f"La probabilidad de obtener los datos dado la hipotesis nula es de: {(boot_medias > media_muestral).mean() + (boot_medias < 0 + (0 - media_muestral)).mean()}")
```
