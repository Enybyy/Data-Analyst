```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

#### Cargar datos

```python
df = pd.read_csv("db/datos_correlacion.txt")
```

```python
X = df['X']
Y = df['Y']
```

```python
plt.figure(figsize=(8,6))
plt.title("Diagrama de dispersión")
plt.xlabel("Variable de entrada")
plt.ylabel("Variable de salida")
plt.scatter(X, Y, 
            edgecolor='blue',
            cmap='viridis',
            linewidths=0.4)
plt.show()
 
```

```python
def crear_linea(b0, b1):
    # Crear línea a partir de la intersección y la pendiente
    x_l = [0, 5]
    y_l = [b0 + b1*x for x in x_l]
    
    return x_l, y_l
```

```python
b0, b1 = 0, 1.4

x_l, y_l = crear_linea(b0, b1)

plt.figure(figsize=(8,6))
plt.title("Diagrama de dispersión")
plt.xlabel("Variable de entrada")
plt.ylabel("Variable de salida")
plt.scatter(X, Y)
plt.plot(x_l, y_l, c="blue")
plt.show()
```
