## HISTOGRAMA

##### Importar Librerías

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

##### Leer data

```python
df_amz = pd.read_csv('data/precios_amazon.csv', sep=';', decimal=',', )
df_eth = pd.read_csv('data/precios_ethereum.csv', sep=';', decimal=',', )
```

```python
print(df_amz.head().to_markdown(),"\n")
print(df_eth.head().to_markdown())
```

##### Función

```python
def histograma(df, name):

    n = len(df)
    k = int(round((1 + 3.3 * np.log(n))))

    plt.figure(figsize=(10,6))
    plt.suptitle(f"{name}", fontsize=16)

    for i, col in enumerate(['Open', 'High', 'Low', 'Close'], start=1):
        ax = plt.subplot(2,2,i)
        ax.hist(df[col], bins=k, edgecolor='black')
        ax.set_title(f"Precio {col}")

        ax.set_xlim(1000, 5000)
        ax.set_ylim(0, 50)

    plt.tight_layout()
    plt.show()
    
```

```python
dataframes = {
    "Histogramas de Amazon" : df_amz,
    "Histogramas de Ethereum" : df_eth,
}

for name, dt in dataframes.items():
    histograma(dt, name)
 
```
