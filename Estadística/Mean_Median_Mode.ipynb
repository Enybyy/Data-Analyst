```python
df_amz = pd.read_csv("data/precios_amazon.csv", sep=";", decimal=",", parse_dates=['Date'], dayfirst=True)
df_eth = pd.read_csv("data/precios_ethereum.csv", sep=";", decimal=",", parse_dates=['Date'])
```

#### CLEAR DATA

```python
df_eth_clean = df_eth.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)
```

```python
df_eth_clean
```

##### MEAN - MEADIAN - MODE

```python
# DF AMAZON - M M M
df_amz_mean = df_amz[['Open', 'High', 'Low', 'Close']].mean()
df_amz_median = df_amz[['Open', 'High', 'Low', 'Close']].median()
df_amz_mode = df_amz[['Open', 'High', 'Low', 'Close']].round().mode().iloc[0]

# DF ETHEREUM - M M M 
df_eth_mean = df_eth_clean[['Open', 'High', 'Low', 'Close']].mean()
df_eth_median = df_eth_clean[['Open', 'High', 'Low', 'Close']].median()
df_eth_mode = df_eth_clean[['Open', 'High', 'Low', 'Close']].round().mode().iloc[0]
```

###### AMAZON

```python
df_amz_stats = pd.DataFrame({
    'Mean': df_amz_mean,
    'Median': df_amz_median,
    'Mode': df_amz_mode,
})

df_amz_stats
```

###### ETHEREUM

```python
df_eth_stats = pd.DataFrame({
    'Mean': df_eth_mean,
    'Median': df_eth_median,
    'Mode': df_eth_mode,
})

df_eth_stats
```

#### FUNCION 

```python
def calcular_estadisticas(df, columnas):
    df_mean = df[columnas].mean()
    df_median = df[columnas].median()
    df_mode = df[columnas].round().mode().iloc[0]

    df_estadisticas = pd.DataFrame({
        'Mean':df_mean,
        'Median':df_median,
        'Mode':df_mode,
    })

    return df_estadisticas
```

```python
dataframes = {
    'Amazon': df_amz,
    'Ethereum': df_eth_clean
}
columns = ['Open', 'High', 'Low', 'Close']

for name, df in dataframes.items():
    estadisticas = calcular_estadisticas(df, columns)
    print(f"Estadisticas de {name}: \n", estadisticas.to_markdown(), "\n")
```
