### Add and calculate a new `Calculated_Date` column

Use `Day`, `Month`, `Year` to create a `Date` column (`YYYY-MM-DD`).

##### Es mas eficiente hacerlo 

```python
#Â your code goes here
sales['Calculated_Date'] = sales[['Year', 'Month', 'Day']].astype(str).agg('-'.join, axis=1)    # Convierte las columnas a texto y concatena en una sola columna de fecha usando una operaciÃ³n vectorizada.
sales.head()                                                                                    # MÃ¡s eficiente que `apply` con `lambda` debido a que evita el procesamiento fila por fila.
```

```python
# And if you try with this
sales['Calculated_Date'] = sales[['Year', 'Month', 'Day']].apply(lambda x: '{}-{}-{}'.format(x.iloc[0], x.iloc[1], x.iloc[2]), axis=1)
sales.head()
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

### How did sales evolve through the years?

Show a <b>line plot</b> using `Calculated_Date` column as the x-axis and the count of sales as the y-axis.

```python
#Â your code goes here
ax = sales['Calculated_Date'].value_counts().sort_index().plot(kind='line', 
                                                          grid=True,
                                                          figsize=(25,6));

# COnfigura el localizador 'principal' del eje x para mostrar 'ticks' en cada aÃ±o.
ax.xaxis.set_major_locator(mdates.YearLocator())
# Configura el localizador 'secundario' del eje x para mostrar 'ticks' el dÃ­a 1 de cada 2 meses.
ax.xaxis.set_minor_locator(mdates.MonthLocator(bymonthday=1, interval=2))
# Formatear el eje principal para mostrar el aÃ±o completo (YYYY)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
# Formatear el eje secundario para mostrar los meses abreviados (MMM)
ax.xaxis.set_minor_formatter(mdates.DateFormatter('%b'))

# AÃ±adir tÃ­tulos
plt.title('Conteo de Ventas por Fecha')
plt.xlabel('Fecha')
plt.ylabel('Conteo de Ventas')

# Mostrar el grÃ¡fico
plt.show()
```
