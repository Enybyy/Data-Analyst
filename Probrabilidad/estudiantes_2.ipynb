#### La columna 'goout' indica con un valor del 1 al 5 la on la que socialisa la persona, siendo 1 'muy bajo' y 5 'muy alto'.
#### Cual es la probabilidad de conseguir una nota mayor o igual a 12 sabiendo que tiene una vida social sana.
#### Se entiende por vida social sana cuando el valor de la columna 'goout' >= 3.

```python
df_prob = df_estudiantes[['goout', 'G3']].copy()
df_prob.head()
```

```python
df_prob.describe()
```

```python
df_prob.hist()
```

```python
df_prob.boxplot()
```

```python
df_prob['condicion_goout'] = np.where((df_prob['goout'] >= 3), 1, 0)
df_prob
```

```python
goout_count = df_prob['condicion_goout'].sum()
print(goout_count)
```

```python
df_prob['condicion_goout_g3'] = np.where((df_prob['goout'] >= 3) & (df_prob['G3'] >=     12), 1, 0)
df_prob
```

```python
goout_g3_count = df_prob['condicion_goout_g3'].sum()
print(goout_g3_count)
```

```python
probabilidad = goout_g3_count/goout_count
print(f"La probabilidad de obtener una nota mayor o igual a 12 teniendo una vida social sana es de: {probabilidad}")
```
