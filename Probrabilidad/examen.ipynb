## PROBABILIDAD BINOMIAL

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
notas = np.random.binomial(20,0.5,10)
notas
```

```python
examenes_aprobados = len([nota for nota in notas if nota >= 11])
print(f"Los exámenes aprobados son: {examenes_aprobados}")
    
```

```python
respuestas_correctas = np.random.binomial(100, 0.5, size=1000)
```

```python
plt.hist(respuestas_correctas, bins=20)
# plt.xlim(0, 100)
# plt.ylim(0, 20)
plt.xlabel('Número de respuestas correctas')
plt.ylabel('Frecuencia')
```
